### 🎬 Leitura frame a frame ( Local e Livestream )

```python
# baixar biblioteca do OpenCV
!pip install cv2
```

```python
# import de bibliotecas

import numpy as np
import cv2 as cv
```

```python
# ler vídeos de uma pasta

captura = cv.VideoCapture('insira vídeo/caminho do vídeo')
 
while captura.isOpened():
    ret, frame = captura.read()
 
    # se o frame for lido corretamente, ret é True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
 
    cv.imshow('frame', gray)
    if cv.waitKey(1) == ord('q'):
        break
 
captura.release()
cv.destroyAllWindows()
```

```python
# captura de vídeos livestream (câmera)

livestream = cv.VideoCapture("fonte_câmera_livestream")
if not livestream.isOpened():
    print("Cannot open camera")
    exit()
while True:
    # Captura frame-by-frame
    ret, frame = livestream.read()
 
    # se o frame for lido corretamente, ret é True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Operação do processamento de frames
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    # dá display no resultado dos frames
    cv.imshow('frame', gray)
    if cv.waitKey(1) == ord('q'):
        break
 
# release da captura
livestream.release()
cv.destroyAllWindows()
```
